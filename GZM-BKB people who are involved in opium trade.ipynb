```python
import pandas as pd

# Specify the file path
file_path = '//Users/zhujiaqi/trifecta/BKB/Dutch Ships and Sailors - Generale Zeemonsterrollen VOC/gzm.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the initial DataFrame and its data types
print("Initial DataFrame:")
print(df)
print("Initial data types:")
print(df.dtypes)

# Example conversion of a specific column, handling NaNs
column_name = ['ZEEVARENDEN EUROPEES', 'ZEEVARENDEN AZIATISCH', 'SOLDATEN EUROPEES', 'SOLDATEN AZIATISCH','PASSAGIERS', 'VOORMAN AZIATISCHE ZEEVARENDEN', 'VOORMANSMAAT AZIATISCHE ZEEVARENDEN', 'MATROZEN AZIATISCHE ZEEVARENDEN','LOON KOK AZIATISCHE ZEEVARENDEN', 'DAS ID HEENREIS', 'DAS ID TERUGREIS']
for col in column_name:
    df[col] = df[col].fillna(0).astype(int)

# Display the updated DataFrame and its data types
print("\nUpdated DataFrame:")
print(df)
print("Updated data types:")
print(df.dtypes)
```
